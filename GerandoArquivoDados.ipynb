# Exemplo de carregamento de um conjunto de dados

## bbc_news_ptbr_summary

https://huggingface.co/datasets/celsowm/bbc_news_ptbr_summary

```python
# Import da biblioteca
from datasets import load_dataset

# Carrega o conjunto de dados
dataset = load_dataset("celsowm/bbc_news_ptbr_summary")

print(dataset)
```

Percorre o conjunto de dados

```python
cont = 1
for i, x in enumerate(dataset["train"]):
      if cont < 10:
        print("\n", x)
        cont = cont + 1
```

### Gerando um arquivo texto (documento.csv) com os dados do dataset usando ";" como separador.

```python
import pandas as pd

# Lista para receber os dados selecionados
dados = []

# Percorre o dataframe
for i, x in enumerate(dataset["train"]):
      # Seleciona somente 40 documentos
      if i <= 40:
        # Adiciona o id e o documento
        dados.append({
            "id": i + 1,
            "documento": x["resumo"]
        })

# Cria o DataFrame para poder salvar os dados
df = pd.DataFrame(dados)

# Salva como CSV
df.to_csv("documentos.csv", index=False, encoding="utf-8", sep=';')
```
